```python
records = transcription_df.to_dict(orient='records')
train = records[:int(0.7*len(records))]
valid = records[int(0.7*len(records)):int(0.9*len(records))]
test  = records[int(0.9*len(records)):]
import json
with open('train.jsonl','w') as file:
    for line in train:
        file.write(json.dumps(line) +'\n\r')
with open('valid.jsonl','w') as file:
    for line in valid:
        file.write(json.dumps(line) +'\n\r')
with open('test.jsonl','w') as file:
    for line in test:
        file.write(json.dumps(line) +'\n\r')
'train',len(train), 'valid', len(valid), 'test', len(test)
```

```python
total = transcription_df.duration.sum()
validt = pd.DataFrame(valid).duration.sum()
traint = pd.DataFrame(train).duration.sum()
testt = pd.DataFrame(test).duration.sum()
'total', total, 'valid', validt, 'train', traint, 'test', testt
```
